```python
import pandas as pd
```

```python
fin_qna_data = pd.read_csv("../financial-qa-dataset.csv")
fin_qna_data.head()
```

```python
#Read metadata file
metadata_ds = pd.read_csv("../metadata.csv")
metadata_ds.head()
```

```python
#Combine Datasets with Metadata
combined_dataset = pd.merge(fin_qna_data, metadata_ds , on ='Document')
combined_dataset.head()
```

```python
#Inspect size
combined_dataset.shape
```

```python
#OPTIONAL : Write file to disk
combined_dataset.to_csv("../financial_qna_with_metadata.csv" , index = False)
```
